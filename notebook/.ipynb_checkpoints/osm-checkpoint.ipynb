```python
import matplotlib.pyplot as plt
import contextily as ctx
# Download the administrative boundary of the 1st district of Budapest
admin_district = ox.geocode_to_gdf('Pankow, Berlin')
admin_poly = admin_district.unary_union
# Download building footprints within the administrative boundary
footprints = ox.features_from_polygon(admin_poly, tags={"building": True})
# Download parks within the administrative boundary
parks = ox.features_from_polygon(admin_poly, tags={'leisure': 'park'})
# Download cafes within the administrative boundary
cafes = ox.features_from_polygon(admin_poly, tags={'amenity': 'cafe'})
# Download additional POI types within the administrative boundary
restaurants = ox.features_from_polygon(admin_poly,
tags={'amenity': 'restaurant'})
schools = ox.features_from_polygon(admin_poly,
tags={'amenity': 'school'})
hospitals = ox.features_from_polygon(admin_poly,
tags={'amenity': 'hospital'})
bus_stops = ox.features_from_polygon(admin_poly,
tags={'highway': 'bus_stop'})
parking_areas = ox.features_from_polygon(admin_poly,
tags={'amenity': 'parking'})
print('Number of restaurants:', len(restaurants))
print('Number of schools:', len(schools))
print('Number of hospitals:', len(hospitals))
print('Number of bus_stops:', len(bus_stops))
print('Number of parking_areas:', len(parking_areas))
```

```python
# Create a plot to visualize the admin boundary and various features
f, ax = plt.subplots(1, 1, figsize=(8, 12))
# Define a poi plotter function
def plot_pois(pois, color, ax, crs, label, zorder):
    pois.to_crs(crs).plot(ax=ax,
                            color=color,
                            markersize=5,
                            alpha=0.8,
                            label=label,
                            zorder=zorder)
# Define the local CRS
crs = 5243


# Plot the administrative boundary
admin_district.to_crs(crs).plot(ax=ax,
                                color='none',
                                edgecolor='k',
                                linewidth=2,
                                label='Administrative Boundary')
# Plot the building footprints
footprints.to_crs(crs).plot(ax=ax,
                        color='midnightblue',
                        alpha=0.7,
                        edgecolor='midnightblue',
                        linewidth=0.5,
                        label='Buildings')
# Plot the parks
parks.to_crs(crs).plot(ax=ax,
                        color='darkgreen',
                        alpha=0.5,
                        edgecolor='darkgreen',
                        linewidth=0.5,
                        label= 'green')
# Plot the pois
plot_pois(cafes, 'red', ax, crs, 'Cafes', zorder=5)
# plot_pois(restaurants, 'blue', ax, crs, 'Resturants', zorder=6)
plot_pois(schools, 'gold', ax, crs, 'Schools', zorder=6)
# plot_pois(hospitals, 'goldenrod', ax, crs, 'Hospitals', zorder=7)
# plot_pois(bus_stops, 'orange', ax, crs, 'Bus Stops', zorder=9)
# plot_pois(parking_areas, 'yellow', ax, crs, 'Parking Areas', zorder=10)
# Add the basemap from contextily
# ctx.add_basemap(ax,
# crs = crs,
# url = ctx.providers.CartoDB.DarkMatterNoLabels)
f.patch.set_facecolor('black')

# Set the axes background color to black
ax.set_facecolor('black')
# Add legend
# Add a legend with all available parameters
# plt.legend(
#     loc='upper right',        # Location of the legend (best, upper right, etc.)
#     bbox_to_anchor=(1, 1),    # Adjust the position of the legend
#     ncol=1,                   # Number of columns (default is 1)
#     title="Legend Title",      # Title for the legend
#     shadow=True,               # Add shadow to the legend box
#     fancybox=True,             # Round corners of the legend box
#     fontsize='medium',         # Font size for the legend text
#     title_fontsize='large',    # Font size for the legend title
#     facecolor='lightgrey',     # Background color of the legend
#     edgecolor='black',         # Edge color of the legend box
#     framealpha=0.8,            # Transparency of the legend box
#     markerfirst=True,          # Whether marker comes first in the legend entry
#     frameon=True               # Display frame around the legend


legend = plt.legend(loc='lower left', facecolor = 'none',edgecolor ='none')

for text in legend.get_texts():
    text.set_color("darkcyan")  # Set font color to red
    
# Add title and axis labels
plt.title('Visualization of Pankow District of Berlin', fontsize=15, color='darkcyan')
# plt.xlabel('Longitude')
# plt.ylabel('Latitude')
ax.axis('off')
# saving the map
NOTEBOOK_PATH = pathlib.Path().resolve()
Map = NOTEBOOK_PATH / "../Map" 

figure_path = os.path.join(Map, "pankow.png")
plt.tight_layout()
plt.savefig(figure_path, dpi=300) #define the size
```
